```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
```

```python
from google.colab import drive
drive.mount('/content/drive')
with open('/content/drive/My Drive/UTS/PidatoPresiden.txt', 'r') as file:
    isi_berita = file.read()
    print(isi_berita)
```

```python
word_tokens = word_tokenize(isi_berita)
print(word_tokens)
```

```python
stop_words = set(stopwords.words('indonesian'))
word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
print(word_tokens_no_stopwords)
```

```python
freq_kata_1 = nltk.FreqDist(word_tokens)
freq_kata_2 = nltk.FreqDist(word_tokens_no_stopwords)
print(freq_kata_1.most_common(20))
print()
print(freq_kata_2.most_common(20))
```

```python
keyword_freq = Counter(word_tokens)
```

```python
# Ambil 10 kata kunci teratas
top_keywords = keyword_freq.most_common(10)

# Pisahkan kata kunci dan frekuensinya
keywords, freq = zip(*top_keywords)

# Buat plot kata kunci
plt.figure(figsize=(12, 6))
plt.bar(keywords, freq)
plt.xlabel('Kata Kunci')
plt.ylabel('Frekuensi')
plt.title('Kata Kunci Teratas dalam Pidato')
plt.xticks(rotation=45)
plt.show()
```

```python
# Membuat Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(keyword_freq)

# Menampilkan Word Cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud Kata Kunci dalam Pidato')
plt.show()
```
